```python
#imports
import numpy as np
import pandas as pd
import math

import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
```

```python
#read in the forceplate data and export it into a csv file
fplate_excel_data = pd.read_excel ("/Users/alexburton/Downloads/ASU_WVB_Force_Plate_Data_Dump.xlsx") 
  
# Write the dataframe object 
# into csv file 
fplate_excel_data.to_csv ("ASU_WVB_Force_Plate_Data_Dump.csv",  
                  index = None, 
                  header=True) 
    
# read csv file and convert  
# into a dataframe object 
fplate_csv_data = pd.DataFrame(pd.read_csv("ASU_WVB_Force_Plate_Data_Dump.csv")) 
  
# show the dataframe 
fplate_csv_data
```

```python

unique_names = fplate_csv_data['Name'].unique()

# Create an empty dictionary to store athlete dataframes
athlete_dataframes = {}

# Iterate over unique names and store each athlete's data in their own DataFrame
# There are many NaN values so I replace them all with 0's. Maybe in the future it would be more beneficial
# to maybe drop the entire row if it has a NaN value in one of the rows we are interested in or fill them with
# a player's average value for each metric so as not to skew any math
# I also make sure each dataframe starts at index 0 because it's necessary for the sliding window function
for name in unique_names:
    athlete_df = fplate_csv_data[fplate_csv_data['Name'] == name].fillna(0).copy()
    athlete_df.reset_index(drop=True, inplace=True)  # Reset index starting from 0
    athlete_dataframes[name] = athlete_df
    
athlete_dataframes
```

```python
"""
Muscle imbalances are a huge factor in volleyball injuries. We should look at predicting injury based on if 
there is imbalance in the left/right jumping or landing. Landing seems especially important due to the force the
knees and legs take. Also Peak Relative Landing Power should be considered. Anything less than -11% shows high risk
of soft tissue injury. Athletes should be operating at >-15% at least. We can look at L/R Impulse along with L/R Braking and 
Landing Power. Knees, ankles, and shoulders are most injuries to happen. 

L|R Avg. Braking Force, L|R Avg. Propulsive Force, L|R Peak Landing Force, L|R Avg. Landing Force, L|R Landing Impulse Index
"""

"""
EX: Feed in data of L|R Avg. Landing Force, predict next Landing Force in sequence. Then analyze if there is a trend to an imbalance one way or the other. We do this for 
whichever player is being looked at
"""

"""
Go through each player and grab the L|R stats and predict the future values for each one. If any of them fall
outside the threshold, flag them for injury
"""

#TODO How do we get the player name based off who's dashboard we are looking at
#For now I just have it iterating through the whole roster

# Set the option to suppress the warning
pd.options.mode.chained_assignment = None  # default='warn'

list = ['Peak Relative Propulsive Power', 'Relative Peak Landing Force', 
                                   'L|R Avg. Braking Force','L|R Avg. Propulsive Force',
                                   'L|R Peak Landing Force','L|R Avg. Landing Force', 'L|R Landing Impulse Index']

for player_name in athlete_dataframes:
    athlete_data = athlete_dataframes[player_name][['Date', 'Peak Relative Propulsive Power', 'Relative Peak Landing Force', 
                                   'L|R Avg. Braking Force','L|R Avg. Propulsive Force',
                                   'L|R Peak Landing Force','L|R Avg. Landing Force', 'L|R Landing Impulse Index']]
    for stat in list:
        """
        This uses a sliding window approach. This model is trained on sets of 3 values from the trianing data and
        predict the next one that would go in that sequence compared to the value directly after the set. The input
        for prediction would then be the last 3 values the player recorded.
        """
        data = athlete_data[stat]
        
        #TODO Figure out optimal window size
        x = []
        y = []
        
        window_size = 3  # Number of historical values to consider

        for i in range(len(data) - window_size):
            # Slice the data to get historical values
            historical_values = data[i:i+window_size]


            # Check if the next value index is within bounds
            next_value_index = i + window_size
            if next_value_index < len(data):
                next_value = data[next_value_index]
            else:
                # Handle the case where the next value index is out of bounds
                next_value = None

    
            x.append(historical_values)  # Historical values
            y.append(next_value)        # Next value

        
        #predict next value
        myseed = 34
        xtr, xte, ytr, yte = train_test_split(x,y,random_state=myseed, test_size=.25)

        # tree with at most nbot bottom nodes
        nbot = 10
        tmod = DecisionTreeRegressor(max_leaf_nodes=nbot)
        tmod.fit(xtr,ytr)

        # look at in-sample fits
        yhat = tmod.predict(xtr)
        ypred = tmod.predict(xte)

        rmse = math.sqrt(mean_squared_error(ypred,yte))

        plt.scatter(ypred,yte,c='black',marker='.')
        plt.plot(yte,yte,c='b')
        plt.title(player_name + " " + stat)
        plt.show()

        # Find the max and min value of y
        max_value = max(y)
        min_value = min(y)
        diff = max_value - min_value

        # rRMSE relative RMSE value for each graph
        # we use rRMSE because each column of data will have separate value ranges and variances so we want
        # to judge it based off that dataset as the regular RMSE values may not be inherently comparable
        rrmse = rmse / diff
        print('rRMSE for decision tree: ',rrmse)
        
        
        #flag column if it has significant deviation 
        #rank ech player by number of columns flagged and how serious the flag is
    


```
