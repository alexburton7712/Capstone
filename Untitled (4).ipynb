```python
#imports
import numpy as np
import pandas as pd
import math

import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

from scipy import stats

import re
```

```python
# For reference, a lot of the pre-processing of this data will not need to be done in PowerBI because you can grab
# individual player columns once the data is imported. All this about reading in .csv files, making dataframes, 
# changing names around, and recombining dataframes to be sorted by player only needs to be done while working
# locally in a non-PowerBI IDE. This whole script starts with the sliding window function in PowerBI.

#read in all data sources
fplate_csv_data_22 = pd.read_csv("./ASU_WVB_Force_Plate_Data_Dump.csv") 
#read in 2022-2023 kinexon data
kinexon_csv_data_22 = pd.read_csv("./Kinexon_Data_Dump.csv") 
#read in 2023-2024 force plate data
fplate_csv_data_23 = pd.read_csv("./ASU VB Force Plate Fall 23 Data.csv") 
#read in 2023-2024 kinexon data
kinexon_csv_data_23 = pd.read_csv("./ASU VB Kinexon Fall 23 Data.csv")
    
#convert each csv file into a pandas dataframe
fplate_dataframe_22 = pd.DataFrame(fplate_csv_data_22)
kinexon_dataframe_22 = pd.DataFrame(kinexon_csv_data_22)
fplate_dataframe_23 = pd.DataFrame(fplate_csv_data_23)
kinexon_dataframe_23 = pd.DataFrame(kinexon_csv_data_23)
```

```python
#Change all 'Name values in the dataframe to be the given ID's
#At the time of writing this the current season is the 2023-2024 season. This part will be obsolete when all
#2022-2023 season players are gone and everyone has id's instead of names

#Dictionary to map each full name to an id
player_dict = {
    'MEADOW COOPER' : 'MC0024', 
    'GELI CYR' : 'GC0012', 
    'CLAIRE JETER' : 'CJ0023', 
    'OLIVIA JOHNSON' : 'OJ0010', 
    'MARTA LEVINSKA' : 'ML004', 
    'MADDIE McLAUGHLIN' : 'MM0025', 
    'ROBERTA RABELO' : 'RR0018', 
    'JADEN RAVNSBORG' : 'JR003', 
    'SHANNON SHIELDS' : 'SS009',
    'Meadow Cooper' : 'MC0024', 
    'Geli Cyr' : 'GC0012', 
    'Claire Jeter' : 'CJ0023', 
    'Olivia Johnson' : 'OJ0010', 
    'Marta Levinska' : 'ML004', 
    'Maddie Mclaughlin' : 'MM0025', 
    'Roberta Rabelo' : 'RR0018', 
    'Jaden Ravnsborg' : 'JR003', 
    'Shannon Shields' : 'SS009'
}

#Replace all 'Name' in the 2022-2023 season values with the corresponding ID
fplate_dataframe_22['Name'] = fplate_dataframe_22['Name'].replace(player_dict)
kinexon_dataframe_22['Name'] = kinexon_dataframe_22['Name'].replace(player_dict)

#List of unique names in the dataframes
unique_names = fplate_dataframe_23['Name'].unique()
print(unique_names)
```

```python
#Combine 2022-2023 and 2023-2024 fplate and kinexon data together
combined_fplate = pd.concat([fplate_dataframe_22, fplate_dataframe_23], ignore_index=True)
combined_kinexon = pd.concat([kinexon_dataframe_22, kinexon_dataframe_23], ignore_index=True)

#Combine both dataframes by name creating a list of dataframes for each player
def combine_dataframes_by_name(df1, df2):
    combined_dataframes = []
    all_names = set(df1['Name']).union(df2['Name'])  # Get all unique names from both dataframes
    
    for name in all_names:
        # Filter data for each athlete from both dataframes
        athlete_data1 = df1[df1['Name'] == name]
        athlete_data2 = df2[df2['Name'] == name]
        
        # Concatenate data for the athlete from both sources
        combined_data = pd.concat([athlete_data1, athlete_data2], ignore_index=True)
        
        # Append the combined dataframe to the list
        combined_dataframes.append(combined_data)
    
    return combined_dataframes

player_dataframes = combine_dataframes_by_name(combined_fplate, combined_kinexon)

#function to get player data by name from the combined_dataframe
def get_player_data_by_name(combined_dataframes, player_name):
    for df in combined_dataframes:
        if player_name in df['Name'].values:
            return df[df['Name'] == player_name]
    return None  # Player not found


window_size = 2  # Number of historical values to consider


#The sliding window function is the loop I wrote dividing the data into x and y values. It looks at a set of 
#window_size values. I first tested it out with window_size = 3 (By the final version of this it might be a 
#different size)so it looks at a total of 4 numbers at a time
#where the first 3 are made into a separate array and added to x. That fourth value is considered the future value
#and sent into the array y. This way the decision tree is deciding one output based off a sequence of numbers.
#That is how the sliding window part works. I then split x and y into train and test sets
#and send them into the decision tree. 
#Essentially the input into the desicion tree is a list of historical values and it is asked to predict the
#next value in the sequence. The input would be a window_size length array and it would output one future value.
def sliding_window(data):
    x = []
    y = []

    for i in range(len(data) - window_size):
        # Slice the data to get historical values
        historical_values = data[i:i+window_size]


        # Check if the next value index is within bounds
        next_value_index = i + window_size
        if next_value_index < len(data):
            next_value = data[next_value_index]
        else:
            # Handle the case where the next value index is out of bounds
            next_value = None


        x.append(historical_values)  # Historical values
        y.append(next_value)        # Next value
    
    return (x,y)
```

```python
"""
This is my old code for a 1 sample t test to see if the prediction is significantly differen than the rest of 
the data. I didn't wanna erase all this in case it could be useful in the future but I switched to using 
the metric of anything predicted over +-15% should be flagged
        #I am gonna use a hypothesis test to compare the actual dataset and the predicted. Might be better to use
        #some other way to see if there is a significant difference between the most recent and predicted. Idk my
        #stats well enough. 
        t_statistic, p_value = stats.ttest_1samp(input, prediction[0])
        
        #sig_diff will be false if the prediction is not significantly different
#         sig_diff = False
#         alpha = 0.05
#         if p_value < alpha:
#             sig_diff = True;
"""       

# Set the option to suppress the warning
pd.options.mode.chained_assignment = None  # default='warn'

#All the metrics about left v right that we can use form the Force Plate data
list = ['L|R Avg. Braking Force','L|R Avg. Propulsive Force',
        'L|R Peak Landing Force','L|R Avg. Landing Force', 'L|R Landing Impulse Index']

#2D array to hold the predicted values for each player
predictions = np.zeros((len(unique_names), len(list)))

#Keeps track of which athlete we are looking at to add into the predicitons array
i = 0

for player_name in unique_names:
    athlete_dataframe = get_player_data_by_name(player_dataframes, player_name)
    athlete_data = athlete_dataframe[[ 'L|R Avg. Braking Force','L|R Avg. Propulsive Force',
                                   'L|R Peak Landing Force','L|R Avg. Landing Force', 'L|R Landing Impulse Index']]
    #list of booleans representing each column of data. A value will be true if there is significant difference
    #between the input and prediction. False otherwise.
    flags = [False, False, False, False, False]

    athlete_data = athlete_data.dropna()
    athlete_data = athlete_data.reset_index(drop=True)
    
    #keeps track of which stat we are looking at to add into predictions array
    j = 0
    for stat in list:
        data = athlete_data[stat]

        x, y = sliding_window(data)
        
        #predict next value
        myseed = 34
        xtr, xte, ytr, yte = train_test_split(x,y,random_state=myseed, test_size=.25)

        # tree with at most nbot bottom nodes
        nbot = 3
        tmod = DecisionTreeRegressor(max_leaf_nodes=nbot)
        tmod.fit(xtr,ytr)

        # look at in-sample fits
        yhat = tmod.predict(xtr)
        ypred = tmod.predict(xte)

        rmse = math.sqrt(mean_squared_error(ypred,yte))

        # Find the max and min value of y
        max_value = max(y)
        min_value = min(y)
        diff = max_value - min_value

        # rRMSE relative RMSE value for each graph
        # we use rRMSE because each column of dataset will have separate value ranges and variances so we want
        # to judge it based off that data as the regular RMSE values may not be inherently comparable
        rrmse = rmse / diff
        
        #Now to get a prediction based off the most recent datapoints.
        input = data[-(window_size):].tolist()
        #Reshape the list to be 2D of this structure: [[value1], [value2], [value3]] like xtr and xte are
        #to input it into the model
        input_2D = [input]
        prediction = tmod.predict(input_2D)
        #save the predictions
        predictions[i][j] = prediction
        #go to the next stat
        j += 1
    #go to the next player
    i += 1
```
