```python
"""
Seems like muscle imbalances in a huge factor in injury. We should look at predicting injury based on if 
there is imbalance in the left/right jumping or landing. Landing seems especially important due to the force the
knees and legs take. Also Peak Relative Landing Power should be considered. Anything less than -11% shows high risk
of soft tissue injury. Athletes should be operating at >-15% at least. We can look at L/R Impulse along with L/R Braking and 
Landing Power. Knees, ankles, and shoulders are most injuries to happen. 

L|R Avg. Braking Force, L|R Avg. Propulsive Force, L|R Peak Landing Force, L|R Avg. Landing Force, L|R Landing Impulse Index
"""


#Decision Tree for kinexon
fplate_csv_copy = fplate_csv_data[['Peak Relative Propulsive Power', 'Relative Peak Landing Force', 'L|R Avg. Braking Force','L|R Avg. Propulsive Force','L|R Peak Landing Force',
                                  'L|R Avg. Landing Force', 'L|R Landing Impulse Index']].copy()

"""
EX: Feed in data of L|R Avg. Landing Force, predict next Landing Force in a certain amount of time in the future:
eg. one week, one month. Then analyze if there is a trend to an imbalance one way or the other
"""

# Remove rows with NaN values
fplate_csv_copy.dropna(inplace=True)


# Visualize the Data
fplate_csv_copy['L|R Avg. Landing Force'].plot()
plt.xlabel('Date')
plt.ylabel('Landing Force')
plt.title('Landing Force Over Time')
plt.show()
x = fplate_csv_copy.drop(columns=['L|R Avg. Landing Force'])
y = fplate_csv_copy['L|R Avg. Landing Force']

### simple train/test split

myseed = 34
xtr, xte, ytr, yte = train_test_split(x,y,random_state=myseed, test_size=.25)

# # tree with at most nbot bottom nodes
nbot = 10
tmod = DecisionTreeRegressor(max_leaf_nodes=nbot)
tmod.fit(xtr,ytr)

## look at in-sample fits
yhat = tmod.predict(xtr)
ypred = tmod.predict(xte)

rmse = math.sqrt(mean_squared_error(ypred,yte))

plt.scatter(ypred,yte,c='black',marker='.')
plt.plot(yte,yte,c='b')
plt.title(f'test rsme with {nbot} bottom nodes: {np.round(rmse,3)}')
plt.show()

#Fidn the max and min value to make the RMSE value normalized
# Find the maximum and minimum values in the 'Total Workload' column
max_value = y.max()
min_value = y.min()
diff = max_value - min_value

# NRMSE
nrmsedt = math.sqrt(mean_squared_error(yte, ypred)) / diff
print('RMSE for decision tree: ',nrmsedt)

# # MAE
# maedt = mean_absolute_error(yte, ypred)
# print('MAE for decision tree: ',maedt)
```
